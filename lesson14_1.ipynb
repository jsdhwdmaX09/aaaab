```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
x = [0, 1, 2]
y = [3, 4, 5]
ax_1.plot(x,y)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
ax1 = figure.add_subplot(1, 3, 1)
ax1.set_title("label1")
ax1 = figure.add_subplot(1, 3, 2)
ax1.set_title("label2")
ax1 = figure.add_subplot(1, 3, 3)
ax1.set_title("label3")

plt.show()
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1, 1, 1)
axes.plot([-1, -4.3, 15, 21, 31])
print(axes.get_xticks())
#print(axes.get_xticklabels()
axes.set_xticklabels(['a','b','c','d','e','f','g','h','i','j','k'])
plt.show()
```
